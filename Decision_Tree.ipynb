```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
X = iris.data[:, 2:] #select rows and 3rd and 4th col. of iris dataset
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth = 2)
tree_clf.fit(X, y)
```

```python
# We can visualize trained decision tree using export_graphviz()
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pydotplus
from IPython.display import Image
import os

# Export the decision tree to a .dot file string
dot_file = export_graphviz(
    tree_clf,
    out_file=None,
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True
)
```

```python
graph = pydotplus.graph_from_dot_data(dot_file)

image = Image(graph.create_png())
display(image)
```
